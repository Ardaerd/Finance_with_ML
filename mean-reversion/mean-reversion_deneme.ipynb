```python
import yfinance as yf
from IPython.display import display
import plotly.graph_objects as go
```

```python
ticker = "BTC-USD"

data = yf.download(ticker, start="2018-01-01", end="2024-12-27")

data.reset_index(inplace=True)
data.columns = [col[0] if col[1] == '' else col[0] for col in data.columns]


columns_to_keep = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
data = data[columns_to_keep]

print(data.columns)
print(data.index)

display(data.head())
display(data.tail())
```

```python
data
```

```python
window_size = 20

data_rolling = data.copy()

data_rolling["mean"] = data_rolling["Adj Close"].rolling(window_size).mean()
data_rolling["std"] = data_rolling["Adj Close"].rolling(window_size).std()

data_rolling = data_rolling.dropna()


display(data_rolling.head(30))
```

```python
k = 2

data_rolling["zscore"] = (data_rolling["Adj Close"] - data_rolling["mean"]) / data_rolling["std"]

# Initialize signal column
data_rolling["signal"] = 0

upper_threshold = data_rolling["mean"] + k * data_rolling["std"]
lower_threshold = data_rolling["mean"] - k * data_rolling["std"]

# Generate buy signals
data_rolling.loc[
    (data_rolling["zscore"] < -k) &
    (data_rolling["Adj Close"] < data_rolling["mean"] - k * data_rolling["std"]),
    "signal"
] = 1

# Generate sell signals
data_rolling.loc[
    (data_rolling["zscore"] > k) &
    (data_rolling["Adj Close"] > data_rolling["mean"] + k * data_rolling["std"]),
    "signal"
] = -1

display(data_rolling)
```

```python
data_rolling["buy_signal"] = data_rolling["Adj Close"].where(data_rolling["signal"] == 1)
data_rolling["sell_signal"] = data_rolling["Adj Close"].where(data_rolling["signal"] == -1)

fig = go.Figure()

# Add the Adj Close price line
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],
    y=data_rolling["Adj Close"],
    mode='lines',
    name='Adj Close',
    line=dict(color='blue')
))

# Add buy signals
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],
    y=data_rolling["buy_signal"],
    mode='markers',
    name='Buy Signal',
    marker=dict(color='green', size=10, symbol='triangle-up')
))

# Add sell signals
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],
    y=data_rolling["sell_signal"],
    mode='markers',
    name='Sell Signal',
    marker=dict(color='red', size=10, symbol='triangle-down')
))

# Add the upper threshold line
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],
    y=upper_threshold,
    mode='lines',
    name='Upper Threshold',
    line=dict(color='orange', width=1)  # Thinner line
))

# Add the lower threshold line
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],
    y=lower_threshold,
    mode='lines',
    name='Lower Threshold',
    line=dict(color='purple', width=1)  # Thinner line
))

# Add layout details
fig.update_layout(
    title="Trading Signals with Upper and Lower Bounds",
    xaxis_title="Date",
    yaxis_title="Price",
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    template='plotly_white'
)

# Show the figure
fig.show()
```

```python
data_rolling.reset_index(drop=True, inplace=True)
```

```python
base_allocation = 0.1  # Start with 10% allocation
data_rolling["dynamic_allocation"] = 0

# Initialize a counter for consecutive buy/sell signals
consecutive_signal = 0

# Iterate through the DataFrame to calculate dynamic allocation
for i in range(1, len(data_rolling)):

    if data_rolling.loc[i, "signal"] == 1 and data_rolling.loc[i-1, "dynamic_allocation"] < 1:  # Buy signal

        if data_rolling.loc[i - 1, "signal"] == 1 and data_rolling.loc[i - 1, "Adj Close"] > data_rolling.loc[i, "Adj Close"]:  # Previous was also a buy
            consecutive_signal += 1

        else:
            consecutive_signal = 1  # Reset counter if switching from sell to buy

        if data_rolling.loc[i - 1, "dynamic_allocation"] + (base_allocation * consecutive_signal) <= 1:
            data_rolling.loc[i, "dynamic_allocation"] = (base_allocation * consecutive_signal) + data_rolling.loc[i - 1, "dynamic_allocation"]
        
        else:
            data_rolling.loc[i, "dynamic_allocation"] = 1

    elif data_rolling.loc[i, "signal"] == -1 and data_rolling.loc[i-1, "dynamic_allocation"] > 0:  # Sell signal

        if data_rolling.loc[i - 1, "signal"] == -1 and data_rolling.loc[i - 1, "Adj Close"] < data_rolling.loc[i, "Adj Close"]:  # Previous was also a sell
            consecutive_signal += 1

        else:
            consecutive_signal = 1  # Reset counter if switching from buy to sell

        if data_rolling.loc[i - 1, "dynamic_allocation"] - (base_allocation * consecutive_signal) >= 0:
            data_rolling.loc[i, "dynamic_allocation"] = data_rolling.loc[i - 1, "dynamic_allocation"] - (base_allocation * consecutive_signal)

        else: 
            data_rolling.loc[i, "dynamic_allocation"] = 0
            
    else:  # Hold signal
        data_rolling.loc[i, "dynamic_allocation"] = data_rolling.loc[i - 1, "dynamic_allocation"]  # Carry forward

# Calculate the position based on dynamic allocation
data_rolling["position"] = data_rolling["dynamic_allocation"] * data_rolling["signal"]
data_rolling["position"] = data_rolling["position"].replace(to_replace=0, method='ffill')

# Calculate returns based on position and price changes
data_rolling["returns"] = data_rolling["position"].shift() * data_rolling["Adj Close"].pct_change()

# Drop rows with NaN values in returns
data_rolling.dropna(subset=["returns"], inplace=True)
```

```python
data_rolling["cum_returns"] = (1 + data_rolling["returns"]).cumprod()
data_rolling["all_in_return"] = (1 + data_rolling["Adj Close"].pct_change()).cumprod()

display(data_rolling.head(50))
```

```python
fig = go.Figure()

# Add the cumulative returns line
fig.add_trace(go.Scatter(
    x=data_rolling["Date"],  # Assuming 'Date' is a column in the DataFrame
    y=data_rolling["cum_returns"],
    mode='lines',
    name='Cumulative Returns',
    line=dict(color='blue')  # Customize the color if needed
))

fig.add_trace(go.Scatter(
    x=data_rolling["Date"], 
    y=data_rolling["all_in_return"],
    mode='lines',
    name='All-In Return',
    line=dict(color='green')
))


fig.update_layout(
    title="Mean Reversion Strategy vs All-In Returns",
    xaxis_title="Date",
    yaxis_title="Cumulative Returns",
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    template='plotly_white'
)

# Show the figure
fig.show()
```
